```python
freqs = Counter(tweets['coordinates'])
#print(str(freqs))
coordinatesDict = {}
for k,v in freqs.items():
    coordinatesDict[k] = [0,0] #first inizialization, [0] is #positive, [1] is #negative

for coordinate,sentiment in zip(tweets['coordinates'],tweets['sentiment']):
    if coordinate != "":
        if sentiment == "positive":
            coordinatesDict[coordinate][0] += 1
        elif sentiment == "negative":
            coordinatesDict[coordinate][1] += 1
#print("----")
#print(str(coordinatesDict))
clean_dict = filter(lambda k: not isnan(k), coordinatesDict)
print(str(coordinatesDict))
```

```python
## Sentiment trading in the time
##

temp = tweets['date'].tolist()
dates = []
for date in temp:
    dates.append(date.split(" ")[0]) #working
print(str(dates))
freqs = Counter(dates)
freqs = sorted(freqs.items())
print("Freqs are:\n")
print(str(freqs))


dateDict = {}
for k,v in freqs:
    dateDict[k] = [0,0] #first inizialization, [0] is #positive, [1] is #negative

for date,sentiment in zip(dates,tweets['sentiment']):
    if date != "":
        if sentiment == "positive":
            dateDict[date][0] += 1
        elif sentiment == "negative":
            dateDict[date][1] += 1
            
print("Dictionary is:\n")
print(str(dateDict))
datesList = []
positiveList = []
negativeList = []
for date,sentiment in dateDict.items():
    #print("Date is: %s and sentiments are %s" % (date,sentiment))
    datesList.append(date)
    positiveList.append(sentiment[0])
    negativeList.append(sentiment[1])
    
```

```python

trace_high = go.Scatter(
                x=datesList,
                y=positiveList,
                name = "Positive Sentiment",
                line = dict(color = 'E60014'),
                opacity = 0.8)

trace_low = go.Scatter(
                x=datesList,
                y=negativeList,
                name = "Negative Sentiment",
                line = dict(color = '#17BECF'),
                opacity = 0.8)

data = [trace_high,trace_low]

layout = dict(
    title = "Sentiment trading"
)

fig = dict(data=data, layout=layout)
py.iplot(fig, filename = "Sentiment trading")
```

```python
## Sentiment trading with hashtags
##

temp = tweets['hashtag'].tolist()
hashtagsList = []
for line in temp:
    hashtags = line.split(" ")
    for hashtag in hashtags:
        hashtagsList.append(hashtag.lower()) #working
#print(str(hashtags))
freqs = Counter(hashtagsList)
#print("Freqs are:\n")
#print(str(freqs))

hashtagsList = []
counterList = []
for k,v in freqs.most_common(20):
    #print("hashtag is: %s with occurrence: %s" % (k,v))
    hashtagsList.insert(0,k)
    counterList.insert(0,v)
    
dist = [
    go.Bar(
        x=counterList,
        y=hashtagsList,
        orientation="h"
)]

#plotly.offline.iplot
py.iplot({"data":dist, "layout":go.Layout(title="Most common hashtags used")})

```

```python
## Sentiment trading on a map
##

locationsUnique = tweets['location']
locationsUnique.fillna('', inplace = True)
#locationsUnique = pd.unique(tweets[['location'].values.ravel('K'))  #location at this point is 
locationsUnique = pd.unique(tweets[['location']].values.ravel('K'))           #already in the format city, state, region

cordinates = tweets['coordinates'].tolist()

print("LocationUnique length is:" + str(len(locationsUnique)))
#print(str(locationsUnique))

locationDict = {}
latDict = {}
lonDict = {}
for k in locationsUnique:
    if k != "":
        locationDict[k] = [0,0,0] #first inizialization, [0] is #positive, [1] is #negative
    
grouped = tweets.groupby(["location", "sentiment"]).size().reset_index(name="Time")
grouped.head(200)

for (idx, row) in grouped.iterrows():
    if row.location != "":
        #print("location is %s and sentiment is %s and coordinate is %s and time is %s" % (row.location, row.sentiment, row.coordinates, row.Time))%%!
        if row.sentiment == "positive":
            locationDict[row.location][0] = row.Time
        elif row.sentiment == "negative":
            locationDict[row.location][1] = row.Time
        elif row.sentiment == "neutral":
            locationDict[row.location][2] = row.Time
        #latLon = row.coordinates.split(",")
        #latDict[row.location] = latLon[0] 
        #lonDict[row.location] = latLon[1]

print("locationDict is:\n")
print(str(locationDict))
```

```python
grouped = grouped.drop(grouped.index[[0,1,2]])
grouped.head(10)
```

```python
#grouped[grouped['location'] =="Badalona,Spain,CT"]
temp = tweets[['location','coordinates']]
temp = temp.drop_duplicates(subset = 'location')
#grouped.head(100)
new = pd.merge(grouped, temp, how='left', on = ['location'], copy=False)
#grouped.merge(temp,on = 'location', how='left', copy=False)
new.head(100)
```

```python
for (idx, row) in new.iterrows():
    if row.location != "":
        #print("location is %s and sentiment is %s and coordinate is %s and time is %s" % (row.location, row.sentiment, row.coordinates, row.Time))
        if row.sentiment == "positive":
            locationDict[row.location][0] = row.Time
        elif row.sentiment == "negative":
            locationDict[row.location][1] = row.Time
        elif row.sentiment == "neutral":
            locationDict[row.location][2] = row.Time
        latLon = row.coordinates.split(",")
        latDict[row.location] = latLon[0] 
        lonDict[row.location] = latLon[1]

#print("locationDict is:\n")
#print(str(locationDict))
# After that, what we have is:
#
# locationDict[location] = [#positive, #negative]
# latDict[location] = latitude
# lonDict[location] = longitute

latList = []
lonList = []
locationList = []
sentimentSize  = []
occurrencesSent = []
for loc,sentArray in locationDict.items():
    if loc != "Agramunt,Spain,CT":
        sentValue = sentArray[0]/(sentArray[0] + sentArray[1] + sentArray[2]) # to modify the weight
        occurrencesSent.append(str(sentArray[0] + sentArray[1] + sentArray[2]))
        #print("SentimentSize for %s is %s" % (loc,sentValue))
        sentimentSize.append(sentValue)
        latList.append(latDict[loc])
        lonList.append(lonDict[loc])
        locationList.append(loc)
    
new['SentimentSize'] = pd.DataFrame(data=sentimentSize)
new['occurrencesSent'] = pd.DataFrame(data=occurrencesSent)
tweets['displayText'] = 'SentimentSize ' + new['SentimentSize'].astype(str) + '<br>' + \
'Occorrence ' + new['occurrencesSent'] + '<br>' + \
'Location ' + new['location'].astype(str)
```
